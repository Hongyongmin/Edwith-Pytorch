```python
import torch
import torch.optim as optim
```

```python
x_train = torch.FloatTensor([[1],[2],[3]]) ##입력데이터
y_train = torch.FloatTensor([[2],[4],[6]]) ##출력데이터

# Linear Regression은 하나의 직선을 찾는 일, W는 기울기(Weight) b 는 y절편(bias)

W = torch.zeros(1, requires_grad=True) #0으로 초기화 한뒤, 학습시킬것을 명시
b = torch.zeros(1, requires_grad=True)

optimizer = optim.SGD([W,b], lr=0.02)

nb_epochs = 1500
for epoch in range(1, nb_epochs + 1):
    hypothesis = x_train *W +b
    cost = torch.mean((hypothesis - y_train) **2) ## Coumpute lost를 구하는 과정
    print(W,b)
    
    optimizer.zero_grad() #미분값 초기화
    cost.backward() #역전파
    optimizer.step() # 파라미터 갱신
```

- optim은 최적화를 간편하게 해주는 모델

- 그 중 SGD는 확률적 경사하강법이라는 것으로, optim의 한 기법

- optimizer.zero_grad()는 텐서들의 미분값을 초기화 해주는 것


zero_grad()로 gradient 초기화, backward()로 계산, step()으로 개선

train 값에 대해서 hypothesis를 예측하고, 차이가 얼만큼 나는지에 대해서 Cost를 계산한 뒤에 Optimizer로 이를 학습하는 과정의 반복





- 실제로 학습시켜 보았더니 W는 2에 가까워지고 b는 0에 가까워지는것을 확인
